$\Large \textbf{Execution Guide for Spyctral}$

$\large \textbf{Import packages}$

$\text{Import spyctral and create the objects}$

```python
import spyctral as spy
```

$\text{Create an object named star\_object.}$  

$\text{These SpectralSummary objects store all the information from a FISA file or a Starlight file.}$

```python
star_object = spy.read_fisa('tests/datasets/case_SC_FISA.fisa')
star_object = spy.read_starlight('tests/datasets/case_SC_Starlight.out')
```

$\newline$

$\Large \textbf{Getting spectra information}$

$ \text{The created object contains several methods to handle the file information. These methods are:}$

$
\text{- header\_info\_df}
$

$
\text{- get\_spectrum(<"spectrum\_name">)}
$

$
\text{- feh\_ratio}
$

$
\text{- get\_all\_properties}
$

$
\text{- plot}
$

$\newline$


$\newline$
$\text{The \emph{header\_info\_df} method returns the information stored in the header attribute.}$

```python
star_object.header_info_df
```

$\newline$

$\newline$
$\text{The information of a single spectrum is obtained with the \emph{get\_spectrum} method.}$

$\text{Starlight files contain three spectra: }\textit{Observed spectrum, Synthetic spectrum, and residual spectrum.}$

```python
star_object.get_spectrum("observed_spectrum")
```

$\newline$

$\text{The \emph{feh\_ratio} method calculates the metallicity ratio [Fe/H] using the instance's Z value and a solar constant (Z\_SUN).}$

```python
star_object.feh_ratio
```

$\newline$

$\text{The \emph{get\_all\_properties} method creates a DataFrame containing all relevant parameters of the instance.}$ 
$\text{Numerical values are formatted to display in scientific notation where necessary.}$

```python
star_object.get_all_properties
```

$\Large \textbf{Plotting the Spectra}$

$\text{The plot method creates a SpectralPlotter object to generate plots from spectral data provided by a SpectralSummary object.}$

$\text{Spectra plotting can be done in several ways.}$

$\text{a. All spectra.}$

$\text{b. Plot a single spectrum.}$

$\text{c. All spectra with an offset spacing.}$

$\text{d. Subplot indivual spectrum.}$

$\newline$

$\large \textbf{Plotting all spectra.}$

```python
star_object.plot()
```

$\large \textbf{A single spectrum can be plotted.}$

```python
star_object.plot.single('observed_spectrum')
```

$\newline$

$\large \textbf{All spectra can be plotted with an offset spacing.}$

```python
star_object.plot.split(2.0)
```

$\newline$

$\large \textbf{Subplotting all spectra.}$

```python
star_object.plot.subplots()
```
